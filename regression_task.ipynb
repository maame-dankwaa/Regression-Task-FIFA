```python
legacy.drop(["player_url", "player_face_url"], axis = 1, inplace = True)
```

```python
d_columns = ["short_name","overall","potential","age",
                "height_cm","weight_kg","weak_foot","skill_moves",
                                "work_rate","pace","shooting","passing","dribbling",
                               "defending","physic","attacking_crossing","attacking_finishing",
                               "attacking_heading_accuracy","attacking_short_passing","attacking_volleys",
                               "skill_dribbling","skill_curve","skill_fk_accuracy","skill_long_passing",
                               "skill_ball_control","movement_acceleration","movement_sprint_speed",
                               "movement_agility","movement_reactions","movement_balance","power_shot_power",
                               "power_jumping","power_stamina","power_strength","power_long_shots","mentality_aggression",
                               "mentality_interceptions","mentality_positioning","mentality_vision","mentality_penalties",
                               "defending_marking_awareness","defending_standing_tackle","defending_sliding_tackle",
                                "goalkeeping_diving","goalkeeping_handling","goalkeeping_kicking","goalkeeping_positioning",
                               "goalkeeping_reflexes"]
```

```python
#creates a dataframe of potentially usable features
legacy_features = pd.DataFrame(legacy[d_columns],columns = d_columns)
```

```python
legacy_features
```

```python
#checks the missing values in each column
legacy_features.isna().sum()
```

```python
#drops rows with missing values if at most 30% of the data is missing values
threshold = 70
minimum_non_missing_values = int((threshold/100)) * legacy_features.shape[0]
```

```python
legacy_features = legacy_features.dropna(thresh=minimum_non_missing_values, axis=1)
```

```python
legacy_features
```

```python
#gets features to be used according to the percentage of nan values of each feature
use = []
no_use = []
for i in legacy_features.columns:
    if ((legacy_features[i].isnull().sum())<(0.3*(legacy_features.shape[0]))):
        use.append(i)
    else:
        no_use.append(i)
```

```python
#classifies usable features into numeric and non-numeric dataframe
legacy_features = legacy_features[use]
numeric_data = legacy_features.select_dtypes(include = np.number)
non_numeric_data =  legacy_features.select_dtypes(include = ["object"])
```

```python
#checks null values in numeric_data dataframe
numeric_data.isnull().sum()
```

```python
#checks null values in non_numeric_data dataframe
non_numeric_data.isnull().sum()
```

```python
#imputes data to replace missing values
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp = IterativeImputer(max_iter=10,random_state=0)
numeric_data = pd.DataFrame(np.round(imp.fit_transform(numeric_data)),columns = numeric_data.columns)
```

```python
numeric_data
```

```python
#encodes work_rate to check its correlation with overall rating
work_rate,actual_values=pd.factorize(non_numeric_data["work_rate"])
```

```python
actual_values
```

```python
work_rate
```

```python
#adds work_rate to numeric_data so its correlation can be checked
numeric_data["work_rate"] = work_rate
```

```python
corr_matrix = numeric_data.corr()
```

```python
#Checks the correlation of all features with the overall rating
corr_matrix["overall"].sort_values(ascending=False)
```

```python
#Features with a correlation greater than 0.4 are used for training and testing
new_columns = ["potential","age","passing","dribbling","physic",
                "attacking_heading_accuracy","attacking_short_passing",
                "skill_curve","skill_long_passing",
                "skill_ball_control","movement_reactions","power_shot_power",
               "power_long_shots", "mentality_vision"]
```

```python
features = pd.DataFrame(numeric_data[new_columns],columns = new_columns)
```

```python
features
```

```python
x = features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
scaled = scale.fit_transform(x)
```

```python
y = numeric_data["overall"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2, random_state = 42, stratify = y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier(n_estimators=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python

gb.fit(xtrain, ytrain)
pkl.dump(gb, open("/Users/maamedankwaaafranie-adjei/Documents/jpn/male_players (legacy).csv" + gb.__class__.__name__ + '.pkl', 'wb'))
y_pred1 = gb.predict(xtest)
```

```python

print(f""" Gradient Boosting:
    Mean Absolute Error: {mean_absolute_error(y_pred1,ytest)},
    Mean Squared Error: {mean_squared_error(y_pred1,ytest,squared=False)}
      """)
```

```python
rf.fit(xtrain, ytrain)
pkl.dump(rf, open("/Users/maamedankwaaafranie-adjei/Documents/jpn/male_players (legacy).csv" + rf.__class__.__name__ + '.pkl', 'wb'))
y_pred2 = rf.predict(xtest)
```

```python
print(f""" Random Forest Classifier:
    Mean Absolute Error: {mean_absolute_error(y_pred2,ytest)},
    Mean Squared Error: {mean_squared_error(y_pred2,ytest,squared=False)}
      """)
```

```python
rf_parameters = {'n_estimators':[50, 100],'max_depth':[None, 10],'min_samples_split':[2, 5]}
```

```python
from sklearn.model_selection import GridSearchCV
gs1=GridSearchCV(rf,param_grid=rf_parameters,scoring="accuracy")
gs1.fit(xtrain, ytrain)
pkl.dump(gs1, open("/Users/maamedankwaaafranie-adjei/Documents/jpn/male_players (legacy).csv" + gs1.__class__.__name__ + '.pkl', 'wb'))
y_pred4 = gs1.predict(xtest)
```

```python
best_model = gs1.best_estimator_
best_parameters = gs1.best_params_
best_model, best_parameters
```

```python
print(f""" Grid Search with Cross Validation(Random Forest Classifier):
    Mean Absolute Error: {mean_absolute_error(y_pred4,ytest)},
    Mean Squared Error: {mean_squared_error(y_pred4,ytest,squared=False)}
      """)
```

```python
from xgboost import XGBRegressor

xgb_regressor = XGBRegressor(random_state=42)
xgb_regressor.fit(xtrain,ytrain)
pkl.dump(xgb_regressor, open("/Users/maamedankwaaafranie-adjei/Documents/jpn/male_players (legacy).csv" + xgb_regressor.__class__.__name__ + '.pkl', 'wb'))
y_pred3 = xgb_regressor.predict(xtest)


xgb_parameters = {
    'n_estimators': [50, 100],
    'max_depth': [4, 5],
    'learning_rate': [0.01, 0.1]}

gs2 = GridSearchCV(XGBRegressor(),xgb_parameters, cv=5) 

gs2.fit(xtrain, ytrain)
pkl.dump(gs2, open("/Users/maamedankwaaafranie-adjei/Documents/jpn/male_players (legacy).csv" + gs2.__class__.__name__ + '.pkl', 'wb'))

best_model = gs2.best_estimator_
best_parameters = gs2.best_params_
```

```python
print(f""" Extreme Gradient Boosting Regressor:
    Mean Absolute Error: {mean_absolute_error(y_pred3,ytest)},
    Mean Squared Error: {mean_squared_error(y_pred3,ytest,squared=False)}
      """)
```
